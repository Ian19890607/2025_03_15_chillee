- 這裡是一個建立`正弦波`圖表的範例

```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸的資料點，從 0 到 2π，共 100 個點
x = np.linspace(0, 2 * np.pi, 100)

# 計算 y 軸的資料點，使用 sin() 函數
y = np.sin(x)

# 使用 plot() 函數繪製正弦波
plt.plot(x, y)


# 設定圖表標題和軸標籤
plt.title('sin plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
# 顯示圖表
plt.show()
```

- 餘弦圖表

```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸的資料點，從 0 到 2π，共 100 個點
x = np.linspace(0, 2 * np.pi, 100)

# 計算 y 軸的資料點，使用 cos() 函數
y = np.cos(x)

# 使用 plot() 函數繪製餘弦波
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('Cosine Wave')
plt.xlabel('x')
plt.ylabel('cos(x)')

# 顯示圖表
plt.show()
```

- 同時顯示sin,cos的正弦波和餘弦波的圖表

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成數據
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# 繪製正弦波和餘弦波
plt.figure(figsize=(10, 5))
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.title('sin and cos')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
