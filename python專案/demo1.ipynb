- 這裡是一個建立`正弦波`圖表的範例

```python
import numpy as np
import matplotlib.pyplot as plt




# 建立 x 軸的資料點，從 0 到 2π，共 100 個點
x = np.linspace(0, 2 * np.pi, 100)

# 計算 y 軸的資料點，使用 sin() 函數
y = np.sin(x)

# 使用 plot() 函數繪製正弦波
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('sin plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
# 顯示圖表
plt.show()
```
